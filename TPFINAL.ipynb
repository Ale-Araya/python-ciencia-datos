```python
# Agrupar los datos por ubicación de la empresa y calcular el salario promedio para cada ubicación
salarios_promedio_por_ubicacion = data.groupby('company_location')['salary_in_usd'].mean().reset_index()

# Ordenar las ubicaciones por salario promedio en orden descendente
salarios_promedio_por_ubicacion = salarios_promedio_por_ubicacion.sort_values(by='salary_in_usd', ascending=False)

# Seleccionar las N ubicaciones principales para graficar
top_n_ubicaciones = 10  # Puedes cambiar este número según sea necesario

# Crear un gráfico de barras para visualizar los salarios promedio por ubicación
plt.figure(figsize=(12, 6))
p = sns.barplot(x='salary_in_usd', y='company_location', data=salarios_promedio_por_ubicacion.head(top_n_ubicaciones), palette='bright')
plt.title('Top {} Salarios Promedio de Data Science por Ubicación'.format(top_n_ubicaciones), fontsize=12, fontweight='bold')
plt.xlabel('Salario Promedio (USD)', fontsize=12, fontweight='bold')
plt.ylabel('Ubicación', fontsize=12, fontweight='bold')

for container in p.containers:
    p.bar_label(container,
                fontsize=12,
                bbox={'boxstyle': 'larrow', 'edgecolor': 'red', 'facecolor': 'white'},
                label_type="edge",
                fontweight='bold'
               )

# Personalizar el color de fondo
plt.gca().set_facecolor("#f4f4f4")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['salary'], orient='horizontal', color='skyblue')
plt.title('Diagrama de Caja para Salarios')
plt.xlabel('Salario (USD)')
plt.show()
```

```python
# Calcular el rango intercuartil (IQR) de la columna 'salary'
Q1 = data['salary'].quantile(0.25)
Q3 = data['salary'].quantile(0.75)
IQR = Q3 - Q1

# Definir los límites inferior y superior para identificar valores atípicos
umbral_inferior = Q1 - 1.5 * IQR
umbral_superior = Q3 + 1.5 * IQR

# Filtrar los datos para mantener solo los valores dentro del rango IQR
data_sin_outliers = data[(data['salary'] >= umbral_inferior) & (data['salary'] <= umbral_superior)]

# Verificar que los valores atípicos se hayan eliminado
print("Datos sin valores atípicos en 'salary':")
print(data_sin_outliers)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=data_sin_outliers['salary'], orient='horizontal', color='skyblue')
plt.title('Diagrama de Caja para Salarios')
plt.xlabel('Salario (USD)')
plt.show()
```

ahora saco la media, despues de sacar los outlayers de salary


```python
# Calcular la media de la columna 'salary' en el DataFrame sin valores atípicos
media_salario_sin_outliers = data_sin_outliers['salary'].mean()

# Imprimir la media del salario sin valores atípicos
print("La media del salario sin valores atípicos es:", media_salario_sin_outliers)
```

```python
# Calcular la matriz de correlación para todas las variables numéricas
matriz_correlacion = data.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriz de Correlación")
plt.show()
```

La tabla muestra una matriz de correlación entre tres variables: 'work_year', 'salary' y 'salary_in_usd'. Quise demostrar que La matriz de correlación es simétrica, y la principalmente siempre seran 1 porque son correlativas.

Correlación entre 'work_year' y 'salary_in_usd' (0.21):

La correlación positiva (0.21) sugiere que hay una relación positiva entre el año de trabajo ('work_year') y el salario en dólares ('salary_in_usd').
Esto significa que a medida que el año de trabajo aumenta, es más probable que el salario en dólares también aumente.
Correlación entre 'work_year' y 'salary' (-0.09):

La correlación negativa (-0.09) indica una relación débil negativa entre el año de trabajo ('work_year') y el salario ('salary').
Esto sugiere que no hay una fuerte relación lineal entre el año de trabajo y el salario en la moneda original ('salary').
La correlación negativa débil podría deberse a la variabilidad en los datos.
Correlación entre 'salary' y 'salary_in_usd' (-0.02):

La correlación cercana a cero (-0.02) indica que no hay una correlación fuerte entre el salario en la moneda original ('salary') y el salario en dólares ('salary_in_usd').
Esto sugiere que el salario en dólares no está fuertemente relacionado con el salario en la moneda original, lo que podría ser resultado de fluctuaciones cambiarias u otras variables que influyen en la conversión.
