```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def generar_datos(N):
    sexo = np.random.choice(['MASCULINO', 'FEMENINO', 'NO ESPECIFICA'], N, p=[0.4, 0.5, 0.1])
    edad = np.clip(np.around(np.random.normal(44, 6, N)), 28, 60)
    altura = np.clip(np.around(np.random.normal(180, 10, N), decimals=1), 150.0, 210.0)
    peso = np.clip(np.around(np.random.normal(80, 15, N), decimals=1), 40.0, 130.0)
    horas_trabajo_estudio = np.clip(np.around(np.random.normal(10, 2, N), decimals=1), 4.0, 12.0)
    horas_ejercicio = np.clip(np.around(np.random.normal(1, 0.5, N), decimals=1), 0, 2)
    horas_descanso = 24 - horas_trabajo_estudio - horas_ejercicio
    datos = np.column_stack((sexo, edad, altura, peso, horas_descanso, horas_trabajo_estudio, horas_ejercicio))
    return datos

N = int(input('Ingrese la cantidad de datos a cargar: '))

datos = generar_datos(N)
#1 Crear un DataFrame para visualizar los datos en forma de tabla
columnas = ['Sexo', 'Edad', 'Altura', 'Peso', 'Horas Descanso', 'Horas Trabajo/Estudio', 'Horas Ejercicio']
df = pd.DataFrame(data=datos, columns=columnas)
print(df)
# Calcular la Media en los Datos de Altura, Peso y Horas
media_altura = np.mean(df['Altura'].astype(float))
media_peso = np.mean(df['Peso'].astype(float))
media_horas_descanso = np.mean(df['Horas Descanso'].astype(float))
media_horas_trabajo_estudio = np.mean(df['Horas Trabajo/Estudio'].astype(float))
media_horas_ejercicio = np.mean(df['Horas Ejercicio'].astype(float))

print(f"Media de Altura: {media_altura:.2f}")
print(f"Media de Peso: {media_peso:.2f}")
print(f"Media de Horas de Descanso: {media_horas_descanso:.2f}")
print(f"Media de Horas de Trabajo/Estudio: {media_horas_trabajo_estudio:.2f}")
print(f"Media de Horas de Ejercicio: {media_horas_ejercicio:.2f}")


# Graficar la Distribución de Alturas y Pesos
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df['Altura'], bins=30, color='blue', alpha=0.7)
plt.xlabel('Altura')
plt.ylabel('Frecuencia')
plt.title('Distribución de Alturas')

plt.subplot(1, 2, 2)
plt.hist(df['Peso'], bins=30, color='green', alpha=0.7)
plt.xlabel('Peso')
plt.ylabel('Frecuencia')
plt.title('Distribución de Pesos')

plt.tight_layout()
plt.show()

# Plotear de Manera Conjunta y Encontrar la Correlación entre Peso y Altura
plt.scatter(df['Altura'], df['Peso'])
plt.xlabel('Altura')
plt.ylabel('Peso')
plt.title('Relación entre Altura y Peso')
plt.show()

# Plotear de Manera Conjunta los Datos de Horas
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(df['Horas Descanso'], df['Horas Trabajo/Estudio'], color='orange', label='Horas Descanso vs. Trabajo/Estudio')
plt.xlabel('Horas de Descanso')
plt.ylabel('Horas de Trabajo/Estudio')
plt.title('Relación entre Horas de Descanso y Trabajo/Estudio')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(df['Horas Descanso'], df['Horas Ejercicio'], color='purple', label='Horas Descanso vs. Ejercicio')
plt.xlabel('Horas de Descanso')
plt.ylabel('Horas de Ejercicio')
plt.title('Relación entre Horas de Descanso y Ejercicio')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#se define una función para generar y limpiar los outliers
def detect_outliers_iqr(data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Llamamos a la función para cada columna numérica donde desees detectar outliers
columns_to_check = ["Age"]
outliers_list = []
for col in columns_to_check:
    outliers = detect_outliers_iqr(df, col)
    outliers_list.append(outliers)

# Concatenamos todos los outliers encontrados en un solo DataFrame
outliers_df = pd.concat(outliers_list)

# Eliminamos los outliers del conjunto de datos original
df_cleaned1 = df.drop(outliers_df.index)

# Imprimimos la cantidad de outliers detectados
print("Número de outliers detectados:", len(outliers_df))
print(outliers)
```
